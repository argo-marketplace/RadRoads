### Notebook Object
This notebook will find:
- longest street
- shortest street  

in Boulder,Colorado, USA

```python
import osmnx as ox
import shapely
import geopandas as gp
import pandas as pd
import os
import matplotlib.pyplot as plt
import fiona
%pylab inline
import numpy as np
```

### 1.1 OSMnx Visualization

```python
Boulder = ox.graph_from_place('Boulder, Colorado, USA')
Boulder_projected = ox.project_graph(Boulder)
fig, ax = ox.plot_graph(Boulder_projected)
```

```python
Boulder_stats = ox.basic_stats(Boulder)
Boulder_stats
```

```python
ox.save_graph_shapefile(Boulder_projected, filename='boulder_osmnx')
```

```python
streets_osmnx = gp.read_file('data/boulder_osmnx/edges/edges.shp')
streets_osmnx.head()
```

### 1.2 Combine Same Street Name

```python
street = pd.DataFrame(streets_osmnx)
```

```python
streets = pd.DataFrame(street.groupby(['name','length'],as_index=False).size())
streets.reset_index(inplace=True)
streets.columns = ['name','length','count']
streets.head()
```

```python
streets.length = streets.length.astype(float)
streets.head()
```

### 1.3 Find Longest Street

```python
street_length = streets.groupby('name')['length'].sum()
street_length = pd.DataFrame(street_length)
street_length.loc[street_length['length'].idxmax()]
```

```python
#find the longest street
street_length.length['Broadway']
```

```python
#street_length.sort_values('length', ascending=False).head()
```

### 1.4 Find Shortest Street

```python
street_length.loc[street_length['length'].idxmin()]
```

```python
#find the shortest street
street_length.length['Streamside Lane']
```

```python
# street_length.sort_values('length', ascending=True).head()
```

### 1.5 Import Nominatim for Street Details

```python
from geopy import Nominatim
```

```python
geolocator = Nominatim()
location = geolocator.geocode("Broadway,Boulder,Colorado")
print(location.address)
print(location.raw)
```

```python
location = geolocator.geocode("Streamside Lane,Boulder,Colorado")
print(location.address)
print(location.raw)
```
